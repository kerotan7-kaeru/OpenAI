```python
# Defaultã€€å…¨ã¦ã®å›ç­”ç”ŸæˆãŒçµ‚ã‚ã£ã¦ã‹ã‚‰ãƒ¬ã‚¹ãƒãƒ³ã‚¹ã‚’è¿”ã•ã›ã‚‹ã€‚å‡ºåŠ›çµæœã‚’ã¾ã¨ã‚ã¦ä¸€æ‹¬ã§è¡¨ç¤ºã™ã‚‹ã€‚

completion = client.chat.completions.create(
  model="gpt-3.5-turbo",
  messages=[
    {"role": "system", "content": "You are a helpful assistant."},
    {"role": "user", "content": "æ—¥æœ¬ã®ãƒãƒ³ã‚·ãƒ§ãƒ³ç®¡ç†æ¥­ã®æ¥­å‹™å†…å®¹ã«ã¤ã„ã¦æ•™ãˆã¦ãã ã•ã„ã€‚"}
   #{"role": "user", "content": "Hello! ã‚ãªãŸã¯èª°ã§ã™ã‹ï¼Ÿ"}
  ]
)

print(completion.choices[0].message)
```

```python
# Streaming  ãƒ–ãƒ©ã‚¦ã‚¶ã®ChatGPTã®ã‚ˆã†ã«é †æ¬¡ãƒ¬ã‚¹ãƒãƒ³ã‚¹ã‚’streamã§è¿”ã•ã›ã‚‹ã€‚å‡ºåŠ›çµæœã¯ãƒˆãƒ¼ã‚¯ãƒ³æ¯ã«è¡¨ç¤ºã™ã‚‹ã€‚

completion = client.chat.completions.create(
  model="gpt-3.5-turbo",
  messages=[
    {"role": "system", "content": "You are a helpful assistant."},
    {"role": "user", "content": "æ—¥æœ¬ã®ãƒãƒ³ã‚·ãƒ§ãƒ³ç®¡ç†æ¥­ã®æ¥­å‹™å†…å®¹ã«ã¤ã„ã¦æ•™ãˆã¦ãã ã•ã„ã€‚"}
  ],
  stream=True  # ğŸ¸â†ã“ã“ã ã‘ãŒç•°ãªã‚‹
)

for chunk in completion:
  print(chunk.choices[0].delta)
```
