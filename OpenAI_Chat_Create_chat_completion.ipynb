```python
# 🐸
# OPENAI_API_KEYの設定
# 事前に、Colabのシークレットに OpenAIの API Keysが設定されている事！

from google.colab import userdata
OPENAI_API_KEY = userdata.get('OpenAI-API_Key')
```

```python
from openai import OpenAI
```

```python
client = OpenAI(
    api_key=OPENAI_API_KEY,
)
```

```python
# Default　全ての回答生成が終わってからレスポンスを返させる。出力結果をまとめて一括で表示する。

completion = client.chat.completions.create(
  model="gpt-3.5-turbo",
  messages=[
    {"role": "system", "content": "You are a helpful assistant."},
    {"role": "user", "content": "日本のマンション管理業の業務内容について教えてください。"}
   #{"role": "user", "content": "Hello! あなたは誰ですか？"}
  ]
)

print(completion.choices[0].message)
```

```python
# Streaming  ブラウザのChatGPTのように順次レスポンスをstreamで返させる。出力結果はトークン毎に表示する。

completion = client.chat.completions.create(
  model="gpt-3.5-turbo",
  messages=[
    {"role": "system", "content": "You are a helpful assistant."},
    {"role": "user", "content": "日本のマンション管理業の業務内容について教えてください。"}
  ],
  stream=True  # 🐸←ここだけが異なる
)

for chunk in completion:
  print(chunk.choices[0].delta)
```
